# Installing the required Libraries

```python
import nrrd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading Raw Data and Visualizing it's dimensions

```python
raw_data = 'research-contributions/Nathan Arnett Calcification/DICOM 12/ESUS 12_7 Unnamed Series.nrrd'

#Reading nrrd file 
data, header = nrrd.read(raw_data)

# Check the dimensions of the loaded image
print(data.shape)  

# View slices along the Z-axis (third dimension)
z_slices = [0, 100, 200]
for z in z_slices:
    plt.imshow(data[:, :, z], cmap='gray')
    plt.title(f'Slice {z}')
    plt.show()
```
